```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure df is loaded from the previous steps, if not already in memory.
df = pd.read_csv('/content/drive/MyDrive/NYPD_ComplaintData.csv')

Borough = ['Bronx', 'Brooklyn', 'Manhattan', 'Queens', 'Staten Island']

# Calculate counts for each Borough from the dataframe
borough_counts = df['BORO_NM'].value_counts().reindex(Borough, fill_value=0)
complaints_counts = df['CMPLNT_NUM'].value_counts().reindex(Borough, fill_value=0)


# Plotting the Borough distribution
plt.barh(borough_counts.index, borough_counts.values, color='skyblue')
plt.barh(complaints_counts.index, complaints_counts.values, color='skyblue')
plt.title('NYPD Complaints by Borough')
plt.xlabel('Number of Complaints')
plt.ylabel('Borough')
plt.show()
```
