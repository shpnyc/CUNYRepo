```python
##Select a school and analyse data by Tendancy --Step Seven;
##Syntax error correction must include a data cleaning procedure to convert data type 'object'
##Assigned the float attribute; int and several others still receiving syntax errors.
##Opened CSV file in Excel to determine cleaning tasks - previous syntax error indicated that a string needed numeric conversion.
##Added print to read CSV
##Added Group By filter for Regents Exam by School DB '84X539'

import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/SchoolData.csv')

selected_columns = df[df['School DBN'] == '84X539']
print(selected_columns)

df = pd.DataFrame({'Mean Score': ['value1', 's', 'value3', '']})

# Replace 's' with pd.NA and convert to nullable string dtype
df['Mean Score'] = df['Mean Score'].replace('s', pd.NA).astype('string')

# Replace empty strings with np.nan (if using a numeric dtype that allows it)
df['Mean Score'] = pd.Series([1, 2, '', 4])
df['Mean Score'] = df['Mean Score'].replace('', np.nan).astype(float)

print(df)
```
